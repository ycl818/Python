```
class Vector:
  def __init__(self,x,y):
    self.x = x
    self.y = y
  def Length(self):
    return (self.x**2+self.y**2)**0.5


p=Vector(3,4)
#會呼叫 Vector.__init(p,3,4)
q=Vector(5,7)
#會呼叫 Vector.__init(q,5,7)
print(p.Length())
print(q.Length())
```

```
class Vector:
  def __init__(self,x,y):
    self.x = x
    self.y = y
  def Length(self):
    return (self.x**2+self.y**2)**0.5
  def __str__(self):
    return 'XD'


p=Vector(3,4)
print(p)
print(str(p))
print(p.__str__())
print(Vector.__str__(p))
#會呼叫 Vector.__init(p,3,4)
q=Vector(5,7)
#會呼叫 Vector.__init(q,5,7)
print(p.Length())
print(q.Length())
```

```
class Vector:
  def __init__(self,x,y):
    self.x = x
    self.y = y
  def Length(self):
    return (self.x**2+self.y**2)**0.5
  def __str__(self):
    return f'({self.x},{self.y})'


p=Vector(3,4)
print(p)
print(str(p))
print(p.__str__())
print(Vector.__str__(p))
```

```
class Vector:
  def __init__(self,x,y):
    self.x = x
    self.y = y
  def Length(self):
    return (self.x**2+self.y**2)**0.5

  def Abs(self):
    return self.Length()

  def __str__(self):
    return f'({self.x},{self.y})'
  def __add__(self,q):
    return Vector(self.x+q.x, self.y+q.y)
  def add(self,q):
    return Vector(self.x+q.x, self.y+q.y)

  def __getitem__(self,i):
    if i==0:
      return self.x
    if i==1:
      return self.y
p=Vector(3,4)
q=Vector(5,7)
print(p)
#會呼叫p.__add__(q)
#會呼叫Vector.__add__(p,q)
print(q)
print(p+q)
print(p.add(q))
print(p.Abs())
print(p[0],p[1])
```
